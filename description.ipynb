```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

def plot_time_series(china_df, japan_df, uk_df, france_df,x='year', y_axis='评论数量'):
    # 创建函数以正确聚合年份数据（确保2014-12-1属于2014年）
    def aggregate_by_year(df):
        df[x] = pd.to_datetime(df[x])
        # 提取年份并按年分组计数
        df['年份'] = df[x].dt.year
        annual_data = df.groupby('年份').size().reset_index(name=y_axis)
        # 将年份转换为日期格式以便绘图
        annual_data[x] = pd.to_datetime(annual_data['年份'].astype(str))
        return annual_data

    # 应用自定义聚合函数
    china_annual = aggregate_by_year(china_df.copy())
    japan_annual = aggregate_by_year(japan_df.copy())
    uk_annual = aggregate_by_year(uk_df.copy())
    france_annual = aggregate_by_year(france_df.copy())

    fig, axs = plt.subplots(4, 1, figsize=(10, 12))

    # 绘制中国数据
    sns.lineplot(data=china_annual, x=x, y=y_axis, ax=axs[0], marker='o', markersize=6)
    china_max = china_annual.loc[china_annual[y_axis].idxmax()]
    axs[0].scatter(china_max[x], china_max[y_axis], color='red', marker='v', s=100)
    axs[0].axvline(x=china_max[x], color='red', linestyle='--', linewidth=0.5)
    axs[0].set_title('中国' + y_axis)
    axs[0].set_xlabel('')  # 移除"上传日期"标签
    axs[0].xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    axs[0].xaxis.set_major_locator(mdates.YearLocator())  # 确保显示每一年

    # 绘制日本数据
    sns.lineplot(data=japan_annual, x=x, y=y_axis, ax=axs[1], marker='o', markersize=6)
    japan_max = japan_annual.loc[japan_annual[y_axis].idxmax()]
    axs[1].scatter(japan_max[x], japan_max[y_axis], color='blue', marker='v', s=100)
    axs[1].axvline(x=japan_max[x], color='blue', linestyle='--', linewidth=0.5)
    axs[1].set_title('日本' + y_axis)
    axs[1].set_xlabel('')  # 移除"上传日期"标签
    axs[1].xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    axs[1].xaxis.set_major_locator(mdates.YearLocator())  # 确保显示每一年

    # 绘制英国数据
    sns.lineplot(data=uk_annual, x=x, y=y_axis, ax=axs[2], marker='o', markersize=6)
    uk_max = uk_annual.loc[uk_annual[y_axis].idxmax()]
    axs[2].scatter(uk_max[x], uk_max[y_axis], color='green', marker='v', s=100)
    axs[2].axvline(x=uk_max[x], color='green', linestyle='--', linewidth=0.5)
    axs[2].set_title('英国' + y_axis)
    axs[2].set_xlabel('')  # 移除"上传日期"标签
    axs[2].xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    axs[2].xaxis.set_major_locator(mdates.YearLocator())  # 确保显示每一年

    # 绘制法国数据
    sns.lineplot(data=france_annual, x=x, y=y_axis, ax=axs[3], marker='o', markersize=6)
    france_max = france_annual.loc[france_annual[y_axis].idxmax()]
    axs[3].scatter(france_max[x], france_max[y_axis], color='purple', marker='v', s=100)
    axs[3].axvline(x=france_max[x], color='purple', linestyle='--', linewidth=0.5)
    axs[3].set_title('法国' + y_axis)
    axs[3].set_xlabel('年份')  # 将最后一个图的标签改为"年份"
    axs[3].xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    axs[3].xaxis.set_major_locator(mdates.YearLocator())  # 确保显示每一年

    plt.tight_layout()
    plt.subplots_adjust(hspace=0.3)  # 调整子图之间的间距

    # 保存图片
    plt.savefig(f'plot_time_series_{y_axis}.png',dpi = 300)
    plt.show()

# 使用示例
plot_time_series(china, japan, uk, france, x = '上传日期', y_axis='视频数量')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

def analyze_language_distribution(df):
    # Sub-function 1: Count the values of the 'language' column
    def count_languages(df):
        return df['Language'].value_counts()

    # Sub-function 2: Plot a pie chart with the largest segment separated
    def plot_language_pie_chart(language_counts):
        # Plot the pie chart
        explode = [0.1 if i == 0 else 0 for i in range(len(language_counts))]
        plt.figure(figsize=(8, 8))
        plt.pie(language_counts, labels=language_counts.index, autopct='%1.1f%%', explode=explode, startangle=140)
        plt.title('Language Distribution')
        plt.legend(title='Languages', loc='upper right')
        plt.show()

    # Call the sub-functions
    language_counts = count_languages(df)
    print(language_counts)
    plot_language_pie_chart(language_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 创建数据框
data = {
    '语言': ['英语', '印地语', '荷兰语', '菲律宾语', '西班牙语', '印度尼西亚语', '法语'],
    '数量': [149060, 25488, 3634, 630, 474, 159, 55]
}
df = pd.DataFrame(data)

# 设置绘图风格和字体
sns.set(style="whitegrid")
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 绘制饼图
fig, ax = plt.subplots(figsize=(12, 8))
wedges, texts, autotexts = ax.pie(df['数量'], labels=None, autopct='', startangle=140, colors=sns.color_palette("pastel"))

# 设置百分比标签，只显示前三名
percentages = [f'{p:.1f}%' for p in 100 * df['数量'] / df['数量'].sum()]
for i, (wedge, percentage) in enumerate(zip(wedges, percentages)):
    if i < 3:  # 只显示前三名的百分比
        angle = (wedge.theta2 - wedge.theta1) / 2. + wedge.theta1
        x = wedge.r * 0.7 * np.cos(np.deg2rad(angle))
        y = wedge.r * 0.7 * np.sin(np.deg2rad(angle))
        ax.text(x, y, f'{df["语言"][i]}: {percentage}', ha='center', va='center', fontsize=12, color='black')

# 添加图例
ax.legend(wedges, df['语言'], title="语言", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.savefig('字幕语言分布.png')

# 显示图形
plt.show()
```

```python
# 导入需要的库
from joypy import joyplot
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

def plot_ridge(china_df, japan_df, uk_df, france_df, field):
    # 01. 准备数据
    china_df['国家'] = '中国'
    japan_df['国家'] = '日本'
    uk_df['国家'] = '英国'
    france_df['国家'] = '法国'

    # Combine all dataframes into one
    combined_df = pd.concat([china_df, japan_df, uk_df, france_df])

    combined_df[f"log_{field}"] = np.log10(combined_df[field] + 1)  # 避免 log(0)
    # combined_df[f"sqrt_{field}"] = np.sqrt(combined_df[field])
    # combined_df[f"boxcox_{field}"], _ = stats.boxcox(combined_df[field] + 1)  # 避免 0 值


    # 02. 绘制图形 - 使用joyplot库
    fig,ax = joyplot(combined_df, by='国家', column=f"log_{field}", figsize= (12,8),
            #传入数据，y轴，x轴，设置图片尺寸
            linecolor="white",# 山脊线的颜色
            colormap=sns.color_palette("coolwarm", as_cmap=True),
            #设置山脊图的填充色，用seaborn库的色盘，选择离散型颜色，as_cmap参数用来更改显示的颜色范围是离散的还是连续的
            background = '#eae0d5');
    # 设置背景色
    fig.set_facecolor('#eae0d5') #设置画布背景色
    # Set title and labels
    plt.title(f'{field} 的分布情况')
    plt.xlabel(f"log10({field} + 1)")
    plt.ylabel('密度')

    # Save the figure
    plt.savefig(f'{field}_ridge_plot.png')
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import ScalarFormatter
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

def plot_density(china_df, japan_df, uk_df, france_df, field):
    # Create a figure with 4 subplots
    fig, axs = plt.subplots(4, 1, figsize=(20, 10), dpi=300)

    # Plot density for China
    sns.kdeplot(china_df[field], ax=axs[0], color='red', fill=True, alpha=0.7)
    axs[0].set_title(f'中国 {field} 的密度图')
    axs[0].set_xlabel(field)
    axs[0].set_ylabel('密度')
    axs[0].xaxis.set_major_formatter(ScalarFormatter(useOffset=False))
    axs[0].ticklabel_format(style='plain', axis='x')

    # Plot density for Japan
    sns.kdeplot(japan_df[field], ax=axs[1], color='blue', fill=True, alpha=0.7)
    axs[1].set_title(f'日本 {field} 的密度图')
    axs[1].set_xlabel(field)
    axs[1].set_ylabel('密度')
    axs[1].xaxis.set_major_formatter(ScalarFormatter(useOffset=False))
    axs[1].ticklabel_format(style='plain', axis='x')

    # Plot density for UK
    sns.kdeplot(uk_df[field], ax=axs[2], color='green', fill=True, alpha=0.7)
    axs[2].set_title(f'英国 {field} 的密度图')
    axs[2].set_xlabel(field)
    axs[2].set_ylabel('密度')
    axs[2].xaxis.set_major_formatter(ScalarFormatter(useOffset=False))
    axs[2].ticklabel_format(style='plain', axis='x')

    # Plot density for France
    sns.kdeplot(france_df[field], ax=axs[3], color='purple', fill=True, alpha=0.7)
    axs[3].set_title(f'法国 {field} 的密度图')
    axs[3].set_xlabel(field)
    axs[3].set_ylabel('密度')
    axs[3].xaxis.set_major_formatter(ScalarFormatter(useOffset=False))
    axs[3].ticklabel_format(style='plain', axis='x')

    # 为每个子图设置横轴从0开始
    for ax in axs:
        ax.set_xlim(0, None)

    # Adjust the spacing between subplots
    plt.tight_layout()
    plt.subplots_adjust(hspace=0.4)

    # Save the figure
    plt.savefig(f'{field}_density_plots.png')
    plt.show()
```

```python
import os
def vlookup_time(c_s_s, china, dir):

    # 合并数据框
    merged_df = pd.merge(c_s_s, china[['视频ID', '上传日期']], left_on='VideoID', right_on='视频ID', how='left')

    # 重命名列
    merged_df.rename(columns={'上传日期': 'Upload_time'}, inplace=True)

    # 删除重复的 '视频ID' 列
    merged_df.drop(columns=['视频ID'], inplace=True)

    # 保存结果
    merged_df.to_csv(os.path.join(dir, 'segmented_subtitles_with_upload_time.csv'), encoding='utf-8-sig', index=False)
```
